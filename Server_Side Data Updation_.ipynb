```python
import flwr as fl

# Define the strategy for server-side aggregation
strategy = fl.server.strategy.FedAvg()

# Start the Flower server
fl.server.start_server(
    server_address="0.0.0.0:8080",  # Server address
    config={"num_rounds": 10},      # Number of federated learning rounds
    strategy=strategy
)
```
